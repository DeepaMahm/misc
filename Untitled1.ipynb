# Trying to solve inverse problem for Lorenz System

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp


def lorenz(t, u, sigma, rho, beta):
    x, y, z = u
    return [sigma * (y - x), x * (rho - z) - y, x * y - beta * z]


sigma_true, rho_true, beta_true = 10.0, 15.0, 8.0 / 3.0
u0 = [-8, 7, 27] #ic
t_span = [0, 1] #taking 0-1 for now
t_eval = np.linspace(t_span[0], t_span[1], 1001)

sol = solve_ivp(lorenz, t_span, u0, args=(sigma_true, rho_true, beta_true), t_eval=t_eval) #generating data not using lorenz data now
u_train = sol.y.T
t_train = sol.t


class model_arch(nn.Module):
    def __init__(self):
        super(model_arch, self).__init__()
        self.fc1 = nn.Linear(1, 50)
        self.fc2 = nn.Linear(50, 50)
        self.fc3 = nn.Linear(50, 50)
        self.fc4 = nn.Linear(50, 50)
        self.fc5 = nn.Linear(50, 3)

    def forward(self, t):
        x = torch.tanh(self.fc1(t))
        x = torch.tanh(self.fc2(x))
        x = torch.tanh(self.fc3(x))
        x = torch.tanh(self.fc4(x))
        x = self.fc5(x)
        return x


net = model_arch()


sigma = torch.tensor([10.0], requires_grad=True)
rho = torch.tensor([15.0], requires_grad=True)
beta = torch.tensor([8.0 / 3.0], requires_grad=True)


def loss_func(t, u, sigma, rho, beta):
    u_pred = net(t)
    x_pred, y_pred, z_pred = u_pred[:, 0], u_pred[:, 1], u_pred[:, 2]

    x_t = torch.autograd.grad(x_pred, t, grad_outputs=torch.ones_like(x_pred), create_graph=True)[0]
    y_t = torch.autograd.grad(y_pred, t, grad_outputs=torch.ones_like(y_pred), create_graph=True)[0]
    z_t = torch.autograd.grad(z_pred, t, grad_outputs=torch.ones_like(z_pred), create_graph=True)[0]

    loss_x = x_t - sigma * (y_pred - x_pred)
    loss_y = y_t - (x_pred * (rho - z_pred) - y_pred)
    loss_z = z_t - (x_pred * y_pred - beta * z_pred)

    loss_data = torch.mean((x_pred - u[:, 0])**2 + (y_pred - u[:, 1])**2 + (z_pred - u[:, 2])**2)

    loss_ic = torch.mean((x_pred[0] - u[0, 0])**2 + (y_pred[0] - u[0, 1])**2 + (z_pred[0] - u[0, 2])**2)

    loss = torch.mean(loss_x**2 + loss_y**2 + loss_z**2) + loss_data + loss_ic
    return loss


optimizer = optim.Adam(list(net.parameters()) + [sigma, rho, beta], lr=0.001)


num_epochs = 10000
for epoch in range(num_epochs):
    t_tensor = torch.tensor(t_train, dtype=torch.float32).view(-1, 1).requires_grad_(True)
    u_tensor = torch.tensor(u_train, dtype=torch.float32)

    optimizer.zero_grad()
    loss = loss_func(t_tensor, u_tensor, sigma, rho, beta)
    loss.backward()
    optimizer.step()

    if epoch % 1000 == 0:
        print(f'Epoch {epoch}, Loss: {loss.item()}, sigma: {sigma.item()}, rho: {rho.item()}, beta: {beta.item()}')


print(f'Final sigma: {sigma.item()}')
print(f'Final rho: {rho.item()}')
print(f'Final beta: {beta.item()}')


t_tensor = torch.tensor(t_eval, dtype=torch.float32).view(-1, 1)
u_pred = net(t_tensor).detach().numpy()
x_pred, y_pred, z_pred = u_pred[:, 0], u_pred[:, 1], u_pred[:, 2]

plt.figure(figsize=(10, 7))
plt.plot(t_eval, u_train[:, 0], label='True x(t)', color='b')
plt.plot(t_eval, x_pred, '--', label='Pred x(t)', color='b')
plt.plot(t_eval, u_train[:, 1], label='True y(t)', color='r')
plt.plot(t_eval, y_pred, '--', label='Pred y(t)', color='r')
plt.plot(t_eval, u_train[:, 2], label='True z(t)', color='g')
plt.plot(t_eval, z_pred, '--', label='Pred z(t)', color='g')
plt.xlabel('Time')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
import numpy as np
import torch
import torch.nn as nn
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

############################# Generating data ##################################
def lorenz_ode(t, y, sigma, rho, beta):
    dydt = [
        sigma * (y[1] - y[0]),
        y[0] * (rho - y[2]) - y[1],
        y[0] * y[1] - beta * y[2]
    ]
    return dydt

t_span = [0, 1]  # Time span
y0 = [1, 0, 0]  # Initial conditions
sigma, rho, beta = 10, 28, 8/3  # Parameters

sol = solve_ivp(lorenz_ode, t_span, y0, args=(sigma, rho, beta), dense_output=True)
t = np.linspace(0, 1, 50)
y_true = sol.sol(t)
t_true = t

t_tensor = torch.tensor(t_true, dtype=torch.float32).view(-1, 1).requires_grad_(True)
y_true_tensor = torch.tensor(y_true.T, dtype=torch.float32)
y_init = torch.tensor(y0, dtype=torch.float32).view(1, -1)

########################## Plotting the true solution ##########################
plt.figure(figsize=(10, 6))
plt.plot(t_true, y_true[0, :], label='True x')
plt.plot(t_true, y_true[1, :], label='True y')
plt.plot(t_true, y_true[2, :], label='True z')
plt.xlabel('Time')
plt.ylabel('State')
plt.legend()
plt.grid(True)
plt.title('Lorenz System: True Solution')
plt.show()

############################## Neural Network ##################################
class ModelArch(nn.Module):
    def __init__(self):
        super(ModelArch, self).__init__()
        self.net = nn.Sequential(
            nn.Linear(1, 64),
            nn.Tanh(),
            nn.Linear(64, 64),
            nn.Tanh(),
            nn.Linear(64, 3)
        )

    def forward(self, t):
        return self.net(t)

model = ModelArch()
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

################################ Loss Function #################################
def loss_func(t, u, u_init, sigma, rho, beta, y_true_tensor):
    u_pred = model(t)
    x_pred, y_pred, z_pred = u_pred[:, 0], u_pred[:, 1], u_pred[:, 2]

    x_t = torch.autograd.grad(x_pred, t, grad_outputs=torch.ones_like(x_pred), create_graph=True)[0]
    y_t = torch.autograd.grad(y_pred, t, grad_outputs=torch.ones_like(y_pred), create_graph=True)[0]
    z_t = torch.autograd.grad(z_pred, t, grad_outputs=torch.ones_like(z_pred), create_graph=True)[0]

    loss_x = x_t - sigma * (y_pred - x_pred)
    loss_y = y_t - (x_pred * (rho - z_pred) - y_pred)
    loss_z = z_t - (x_pred * y_pred - beta * z_pred)

    loss_obs = criterion(u_pred, y_true_tensor)

    loss_ic = torch.mean((x_pred[0] - u_init[0, 0])**2 + (y_pred[0] - u_init[0, 1])**2 + (z_pred[0] - u_init[0, 2])**2)

    loss_ode = torch.mean(loss_x**2 + loss_y**2 + loss_z**2)

    lambda1 = 1e-4
    lambda2 = 1
    lambda3 = 1e-2

    loss = lambda1 * loss_ode + lambda2 * loss_obs + lambda3 * loss_ic
    return loss, loss_ic, loss_obs, loss_ode, x_t, y_t, z_t, loss_x, loss_y, loss_z

################################ Training loop #################################
epochs = 1
loss_ic_list = []
loss_obs_list = []
loss_ode_list = []
loss_total_list = []

for epoch in range(epochs):
    # optimizer.zero_grad()  # Moved before model prediction for consistency

    loss, loss_ic, loss_obs, loss_ode, x_t, y_t, z_t, loss_x, loss_y, loss_z = loss_func(t_tensor, y_true_tensor, y_init, sigma, rho, beta, y_true_tensor)
    # y_pred = model(t_tensor.unsqueeze(1))
    # loss = criterion(y_pred, y_true_tensor)

    # print(f"x_t: {x_t}")
    # print(f"y_t: {y_t}")
    # print(f"z_t: {z_t}")
    # print(f"loss_x: {loss_x}")
    # print(f"loss_y: {loss_y}")
    # print(f"loss_z: {loss_z}")


    if torch.isnan(loss):
        print(f'NaN loss encountered at epoch {epoch+1}. Exiting training loop.')
        break
    optimizer.zero_grad()
    loss_ic.backward()
    loss_obs.backward()
    loss_ode.backward()
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 100 == 0 or epoch == 0:
        loss_ic_list.append(loss_ic.item())
        loss_obs_list.append(loss_obs.item())
        loss_ode_list.append(loss_ode.item())
        loss_total_list.append(loss.item())

    if (epoch + 1) % 1000 == 0 or epoch == 0:
        print(f"Epoch: {epoch+1}")
        print(f"Observational Loss: {loss_obs.item()}")
        print(f"IC Loss: {loss_ic.item()}")
        print(f"ODE Loss: {loss_ode.item()}")
        print(f"Total Loss: {loss.item()}")

        with torch.no_grad():
            y_pred = model(t_tensor).numpy()
        plt.figure(figsize=(10, 6))
        plt.plot(t, y_true[0, :], label='True x')
        plt.plot(t, y_true[1, :], label='True y')
        plt.plot(t, y_true[2, :], label='True z')
        plt.plot(t, y_pred[:, 0], '--', label='Predicted x')
        plt.plot(t, y_pred[:, 1], '--', label='Predicted y')
        plt.plot(t, y_pred[:, 2], '--', label='Predicted z')
        plt.xlabel('Time')
        plt.ylabel('State')
        plt.title(f'Lorenz System: True vs Predicted, Epoch {epoch+1}')
        plt.legend()
        plt.grid(True)
        plt.show()

########################## Ploting the final results ###########################
with torch.no_grad():
    y_pred = model(t_tensor).numpy()

plt.figure(figsize=(10, 6))
plt.plot(t, y_true[0, :], label='True x')
plt.plot(t, y_true[1, :], label='True y')
plt.plot(t, y_true[2, :], label='True z')
plt.plot(t, y_pred[:, 0], '--', label='Predicted x')
plt.plot(t, y_pred[:, 1], '--', label='Predicted y')
plt.plot(t, y_pred[:, 2], '--', label='Predicted z')
plt.xlabel('Time')
plt.ylabel('State')
plt.title('Lorenz System: True vs Predicted')
plt.legend()
plt.grid(True)
plt.show()



############################ Ploting the losses ################################
epochs_range = range(0, (epochs // 100) * 100 + 1, 100)

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(epochs_range, loss_ic_list, label='IC Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('IC Loss vs Epochs')
plt.legend()
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(epochs_range, loss_obs_list, label='Observational Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Observational Loss vs Epochs')
plt.legend()
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(epochs_range, loss_ode_list, label='ODE Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('ODE Loss vs Epochs')
plt.legend()
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(epochs_range, loss_total_list, label='Total Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Total Loss vs Epochs')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
